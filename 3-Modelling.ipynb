# Modelling - building a recommendation engine

```python
import pandas as pd
import numpy as np

import pickle 

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Import Warnings
import warnings
warnings.filterwarnings("ignore")
```

## Load data

```python
data = pickle.load(open("../Data/steam/alldata.pkl", 'rb'))
```

```python
data.head()
```

```python
# Get relevant columns
recdata = data[['steam_id', 'id']]
```

```python
recdata.head()
```

```python
# Add binary owned column
recdata['owned'] = np.ones(shape = recdata.shape[0])
recdata.head()
```

```python
# Save using pickle
pickle.dump(recdata, open('recdata.pkl','wb'))
```

```python
# Pivot table
user_id = recdata.pivot_table(index = ['steam_id'], columns = ['id'], values = 'owned')
user_id.head()
```
